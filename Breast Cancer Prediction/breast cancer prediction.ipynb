# Remove the unwanted column

```python
df = df.drop(columns = ['id', 'Unnamed: 32'], axis = 1)
```

# Target column

```python
df['diagnosis'].value_counts()
```

# Convert the string column to numerical column

```python
df['diagnosis'] = df['diagnosis'].map({'M' : 1, 'B' : 0})
```

```python
df.head()
```

# Visualizing the target class

```python
df['diagnosis'].value_counts().plot(kind = 'pie')
plt.title('Class Distribution')
```

# Store the independent feature in 'x' & dependent feature in 'y'

```python
x = df.drop(columns = ['diagnosis'], axis =1)
y = df['diagnosis']
```

# Train/Test split
- Split data into two part : a training set and a testing set
- Train the model(s) on training set
- Test the Model(s) on Testing set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

# Loading the Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

# Training the Model

```python
log = LogisticRegression()
log.fit(x_train, y_train)

svc = SVC()
svc.fit(x_train, y_train)

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)

rf = RandomForestClassifier()
rf.fit(x_train, y_train)

gb = GradientBoostingClassifier()
gb.fit(x_train, y_train)
```

# Prediction on test data

```python
y_pred1 = log.predict(x_test)
y_pred2 = svc.predict(x_test)
y_pred3 = dt.predict(x_test)
y_pred4 = rf.predict(x_test)
y_pred5 = gb.predict(x_test)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
acc1 = accuracy_score(y_test, y_pred1)
acc2 = accuracy_score(y_test, y_pred2)
acc3 = accuracy_score(y_test, y_pred3)
acc4 = accuracy_score(y_test, y_pred4)
acc5 = accuracy_score(y_test, y_pred5)
```

```python
final_data = pd.DataFrame({'Models':['LR','SVM','DT','RF','GB'],
                          'ACC':[acc1*100, acc2*100,acc3*100, acc4*100, acc5*100]})
```

```python
final_data
```

```python
sns.barplot(x = 'Models', y = 'ACC', data = final_data)
```
