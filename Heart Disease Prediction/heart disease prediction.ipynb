# Store the independent features in 'x' & dependent features in 'y'

```python
x = df.drop(columns = ['target'], axis = 1)
y = df['target']
```

```python
x
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(x)
```

```python
X
```

# Train/Test split
- Split data into two part : a training set and a testing set
- Train the model(s) on training set
- Test the Model(s) on Testing set

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
y_pred1 = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc1 = accuracy_score(y_test, y_pred1)
print(acc1)
```

# SVC

```python
from sklearn.svm import SVC
svm = SVC()
```

```python
svm.fit(x_train,y_train)
y_pred2 = svm.predict(x_test)
```

```python
acc2 = accuracy_score(y_test, y_pred2)
print(acc2)
```

# KNeighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier()
```

```python
KNN.fit(x_train, y_train)
y_pred = KNN.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
# tuning the no.of.neighbors

score = []

for k in range(1,41):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    score.append(accuracy_score(y_test,y_pred))
    
print(score)
```

```python
max(score)
```

```python
# to find the index position of the maximum value in the list
score.index(max(score))
```

```python
plt.plot(score)
plt.xlabel("K Value")
plt.ylabel("Acc")
plt.show()
```

```python
Knn=KNeighborsClassifier(n_neighbors=19) 

Knn.fit(x_train, y_train)

y_pred3=knn.predict(x_test)

acc3 = accuracy_score(y_test, y_pred3)

print(acc3)
```
