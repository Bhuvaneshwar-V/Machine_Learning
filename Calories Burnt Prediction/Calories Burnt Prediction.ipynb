# Importing the necessary libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Loading the dataset

```python
exercise_df = pd.read_csv('exercise.csv')
```

```python
# first five rows of the dataset
exercise_df.head()
```

```python
calories_df = pd.read_csv('calories.csv')
```

```python
# first five rows of the dataset
calories_df.head()
```

# Combining two dataframes

```python
df = pd.merge(exercise_df, calories_df, on='User_ID', how='inner')
```

```python
# first 5 rows of the dataset
df.head()
```

```python
# last 5 rows of the dataset
df.tail()
```

```python
# shape of the dataset
df.shape
```

```python
# number of rows and columns
print("Number of rows", df.shape[0])
print("Number of columns", df.shape[1])
```

```python
# basic information about the data
df.info()
```

```python
# checking for null values
df.isnull().sum()
```

```python
# statistical information about the dataset
df.describe()
```

# Data Visualization

```python
# gender column
df['Gender'].value_counts()
```

```python
# plotting the gender column
df['Gender'].value_counts().plot(kind = 'bar')
plt.ylabel('Counts')
plt.xticks(rotation = 0)
plt.show()
```

```python
# dropping the unwanted column
df = df.drop(columns = ['User_ID'], axis = 1)
```

# Distribution of the numerical column

```python
for col in df.columns:
    if df[col].dtype != object:
        sns.distplot(df[col])
        plt.title(col)
        plt.show()
```

```python
# converting the text data into numerical data
df['Gender'] = df['Gender'].map({'female':0, 'male':1})
```

```python
df.head()
```

# Separating features & target

```python
X = df.drop(columns = ['Calories'], axis = 1)
Y = df['Calories']
```

```python
print(X)
```

```python
print(Y)
```

# Splitting the data into train & test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

```python
print(X.shape, x_train.shape, x_test.shape)

print(Y.shape, y_train.shape, y_test.shape)
```

# XGBoost Regressor

```python
from xgboost import XGBRegressor

model = XGBRegressor()
```

```python
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```

# Evaluation

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
score = r2_score(y_test, y_pred)
print('R2_score', score)
```

```python
mse = mean_squared_error(y_test, y_pred)
print('MSE', mse)
```
